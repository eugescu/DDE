```python
import numpy as np
import itertools
import numpy as np
import math
from matplotlib import pyplot as plt
import random

```

```python
times = np.linspace(0.0, 1e3, num = int(1e4))
x_t = [random.random() for t in times]     # time series data
plt.figure(figsize=(15,5))
plt.plot(times, x_t, '.')
plt.show()
```

```python
plt.plot(np.correlate(x_t, x_t, mode='same'))
plt.show()



plt.plot(np.correlate(x_t, x_t, mode='full'))
plt.show()



plt.plot(np.correlate(x_t, x_t, mode='valid'))
plt.show()

```

```python
def estimated_autocorrelation(x):
    n = len(x)
    variance = np.var(x)
    x = x-np.mean(x)
    r = np.correlate(x, x, mode = 'same')[-n:]
    result = r/(variance*(np.arange(n, 0, -1)))
    return result



plt.plot(estimated_autocorrelation(x_t))
plt.show()

plt.plot(estimated_autocorrelation(x_t)[int(len(x_t)-100):int(len(x_t)+100)])
plt.show()
```

```python
import numpy as np
```

```python
# detuning range
N_w = 31
w_r = 0.1
x_range = np.linspace(-w_r,w_r,N_w)
y_range = np.linspace(-w_r,w_r,N_w)

pin = [0] * N_w
pout = [0] * N_w

pss = [[0] * N_w] * N_w
frq = np.zeros((N_w,N_w))

for x_i in range(N_w):
    for y_i in range(N_w):
        w0 = x_range[x_i]
        w1 = y_range[::-1][y_i]

        frq[x_i, y_i] = w0 * w1
        pss[x_i][y_i] = (x_i, y_i)
        pin[y_i] = y_i
    pout[x_i] = [(x_i,p) for p in pin]

    
    
frq, pss
psss = [[(x_i, y_i) for y_i in range(N_w)] for x_i in range(N_w)]
```
